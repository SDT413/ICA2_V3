```python
# importing the library that will allow author to implement dataset and do the data cleaning
import pandas as pd
import matplotlib as plt
```

```python
#importing the Coronavirus tables and cutting off the data which is not needed for the project graphs
#Author indicated columns by writing them after read_csv command
COVID_2022 = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/12-05-2022.csv')[['Lat','Long_','Country_Region', 'Confirmed', 'Deaths','Recovered']]
COVID_2022_copy = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/12-05-2022.csv')[['Lat','Long_','Country_Region', 'Confirmed', 'Deaths','Recovered']]
#Checking if Coronavirus table 2022 data has NAN values because they will have no use on map or graphs
#Checking if table has NAN values
if  COVID_2022.isnull().values.any():
    print("NAN values were found")
    #In order to use rows with NAN values, instead of deleting them, author replaces them with "No province" which will allowed to group them by provinces
    for i in COVID_2022:
        if COVID_2022[i].isnull().values.any():
            # For every Column NAN value must be replaced with its unique value
            if COVID_2022[i].name == "Confirmed":
                COVID_2022.Confirmed =  COVID_2022.Confirmed.fillna(0)
            if COVID_2022[i].name == "Deaths":
                COVID_2022.Deaths =  COVID_2022.Deaths.fillna(0)
            if COVID_2022[i].name == "Recovered":
                COVID_2022.Recovered = COVID_2022.Recovered.fillna(0)
        if COVID_2022[i].sum() == 0:
           del COVID_2022[i]
    print("All NAN values were replaced by status")
    #Distribution check
    COVID_2022['Deaths'].plot(kind='kde')
    COVID_2022_copy['Deaths'].plot(kind='kde')
    #Checking if there are columns that are field with 0
    #As resault it is seen that nothing changed from the moment of Nan value replacement
else:
    print("Data has no NAN values")
```
